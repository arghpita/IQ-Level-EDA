```python
sns.boxplot(df['education_expenditure'])
```

```python
q1=df['education_expenditure'].quantile(0.25)
q3=df['education_expenditure'].quantile(0.75)
```

```python
IQR=q3-q1
lower=q1-1.5*IQR
upper=q3+1.5*IQR
```

```python
df=df[(df['education_expenditure']>lower) & (df['education_expenditure']<upper)]
df
```

```python
sns.boxplot(df['education_expenditure'])
```

```python
sns.boxplot(df['avg_income'])
```

```python
q1=df['avg_income'].quantile(0.25)
q3=df['avg_income'].quantile(0.75)
IQR=q3-q1
lower=q1-1.5*IQR
upper=q3+1.5*IQR
df=df[(df['avg_income']>lower) & (df['avg_income']<upper)]
df
```

```python
sns.boxplot(df['avg_income'])
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['country'],y=df['IQ'],palette='pastel')
plt.title('Distribution of IQ worldwide')
plt.xlabel("Countries")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['country'],y=df['education_expenditure'], palette='deep')
plt.xticks(rotation='vertical')
plt.xlabel('Countries')
plt.ylabel('Education Expenditure')
plt.title('Distribution of Education Expenditure Worldwide')
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['country'], y=df['avg_income'], palette='muted')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Average Income')
plt.title('Distribution of Average Income Worldwide')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 5), sharey=True)

#IQ
sns.histplot(ax=axes[0], x=df['IQ'])
axes[0].set_title('IQ')

#Average Income
sns.histplot(ax=axes[1], x=df['avg_income'])
axes[1].set_title('Average Income')

#Education Expenditure
sns.histplot(ax=axes[2], x=df['education_expenditure'])
axes[2].set_title('Education Expenditure')

#Average Temp
sns.histplot(ax=axes[3], x=df['avg_temp'])
axes[3].set_title('Average Temp')
```

# TOP 10 Countries with Highest Average Income

```python
highest_avg_income=df[['country','avg_income']]
highest_avg_income
```

```python
highest_avg_income=highest_avg_income.sort_values(by='avg_income', ascending=False)[:10]
highest_avg_income
```

```python
plt.figure(figsize=(5,3))
sns.barplot(x=highest_avg_income['country'],y=highest_avg_income['avg_income'],palette='pastel')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Average Income')
plt.title('Top 10 Countries with Highest Average Income')
```
